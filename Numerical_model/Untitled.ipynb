```python
def add_dsqr(seed,xoo,yoo):
    np.random.seed(seed)
    a = np.random.randn(len(xoo))/3
    ao = 0.3*a
    x1 = xoo-ao/np.sqrt(2)
    y1 = ao/np.sqrt(2)+yoo
    plt.scatter(x1,y1)
    return x1,y1
```
