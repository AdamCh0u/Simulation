```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x,y1,y2 = [],[],[]
for c in np.logspace(-2,1,40):
    a = []
    n = []
    for k in range(10):
        demo = city(100,c,10000)
        index = demo.step()
        fit = index.curveFit(100,10)
        a.append(fit[0])
        n.append(fit[1])
        print(k)
    avga = sum(a)/10
    avgn = sum(n)/10
    x.append(c)
    y1.append(avga)
    y2.append(avgn)
    print("======="+str(c)+"=======")
plt.plot(x,y1)
plt.plot(x,y2)
```
