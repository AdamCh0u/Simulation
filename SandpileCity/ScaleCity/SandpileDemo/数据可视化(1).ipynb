# Histogram

A histogram is an accurate representation of the distribution of numerical data. It is an estimate of the probability distribution of a continuous variable (quantitative variable) and was first introduced by Karl Pearson.[1] It is a kind of bar graph. To construct a histogram, the first step is to "bin" the range of values—that is, divide the entire range of values into a series of intervals—and then count how many values fall into each interval. The bins are usually specified as consecutive, non-overlapping intervals of a variable. The bins (intervals) must be adjacent, and are often (but are not required to be) of equal size.

__解释:直方图是用来衡量连续变量的概率分布的。在构建直方图之前，我们需要先定义好bin（值的范围），也就是说我们需要先把连续值划分成不同等份，然后计算每一份里面数据的数量。__


在数据可视化的过程中，图片中的文字经常被用来注释图中的一些特征。使用annotate()方法可以很方便地添加此类注释。在使用annotate时，要考虑两个点的坐标：被注释的地方xy(x, y)和插入文本的地方xytext(x, y)

````python
X = np.linspace(0, 2*np.pi,100)# 均匀的划分数据
Y = np.sin(X)
Y1 = np.cos(X)

plt.plot(X,Y)
plt.plot(X,Y1)
plt.annotate('Points',
         xy=(1, np.sin(1)),#被注释的地方：1，sin1的位置
         xytext=(2, 0.5), fontsize=16,#文本位置，文本大小
         arrowprops=dict(arrowstyle="->"))#箭头参数
plt.title("这是一副测试图！")
````

绘制子图

````python
import numpy as np
import matplotlib.pyplot as plt
#import seaborn as sns

np.random.seed(19680801)

n_bins = 10
x = np.random.randn(1000, 3)#1000行3列

fig, axes = plt.subplots(nrows=2, ncols=2,facecolor='darkslategray') #创建2行2列画布，返回fig和axes
ax0, ax1, ax2, ax3 = axes.flatten()#左上左下右上右下 提取位置，指代子图

colors = ['red', 'tan', 'lime']
ax0.hist(x, n_bins, normed=1, histtype='bar', color=colors, label=colors)
ax0.legend(prop={'size': 10})
ax0.set_title('bars with legend')

ax1.hist(x, n_bins, normed=1, histtype='bar', stacked=True)
ax1.set_title('stacked bar')

ax2.hist(x, n_bins, histtype='step', stacked=True, fill=False)
ax2.set_title('stack step (unfilled)')

# Make a multiple-histogram of data-sets with different length.
# x_multi = [np.random.randn(n) for n in [10000, 5000, 2000]]
# ax3.hist(x_multi, n_bins, histtype='bar')
# ax3.set_title('different sample sizes')

fig.tight_layout() # Adjust subplot parameters to give specified padding.
plt.show()
````

````python
# ShareX or ShareY
N_points = 100000
n_bins = 20

# Generate a normal distribution, center at x=0 and y=5
x = np.random.randn(N_points)
y = .4 * x + np.random.randn(100000) + 5

fig, axs = plt.subplots(1, 2, sharey=True, tight_layout=True)#共享y

# We can set the number of bins with the `bins` kwarg
axs[0].hist(x, bins=n_bins)
axs[1].hist(y, bins=n_bins)
````

## Visualizing linear relationships  

输入特征与输出特征相关性  

统计上的解释  ：共现性，根据输入之间的相关性去掉一个

````python
sns.lmplot(x="total_bill", y="tip", data=tips)#lmplot，两个连续变量相关性 阴影部分是置信区间
````

````python
anscombe = sns.load_dataset("anscombe")
sns.lmplot(x="x", y="y", data=anscombe.query("dataset == 'I'"),
            scatter_kws={"s": 80});#
````

````python
sns.lmplot(x="x", y="y", data=anscombe.query("dataset == 'II'"),
           ci=None, scatter_kws={"s": 80});
````

````python
sns.lmplot(x="x", y="y", data=anscombe.query("dataset == 'II'"),
           order=2, ci=None, scatter_kws={"s": 80});#二次拟合，非线性拟合，order 次数为2
````

````python
sns.lmplot(x="x", y="y", data=anscombe.query("dataset == 'III'"),
           ci=None, scatter_kws={"s": 80});#剔除异常拟合，鲁棒性
````

````python
sns.lmplot(x="x", y="y", data=anscombe.query("dataset == 'III'"),
           robust=True, ci=None, scatter_kws={"s": 80});#鲁棒性
````
